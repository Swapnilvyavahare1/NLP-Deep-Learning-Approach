```python
import zipfile
zip_ref= zipfile.ZipFile('/content/googlenewsvectorsnegative300.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
# Library for word2vec
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
# Pretrained model
model= KeyedVectors.load_word2vec_format('/content/GoogleNews-vectors-negative300.bin.gz', binary=True, limit= None)
```

```python
model['cricket']
```

```python
model['tennis']
```

```python
model.most_similar('tennis')
```

```python
model.similarity('man','woman')
```

```python
model.similarity('man','PHP')
```

```python
model.doesnt_match(['PHP','java','monkey'])
```
